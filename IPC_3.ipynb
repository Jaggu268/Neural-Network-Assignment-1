```python
# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Check for null values in the DataFrame
print("Null values in each column:")
print(df.isnull().sum())

# Replace null values with the mean of each column
df.fillna(df.mean(), inplace=True)

# Verify that there are no more null values
print("\nNull values after replacing with mean:")
print(df.isnull().sum())
```

```python
# Select columns for aggregation
columns_to_aggregate = ['Duration', 'Pulse']

# Aggregate data
aggregation = df[columns_to_aggregate].agg(['min', 'max', 'count', 'mean'])

print("Aggregated Data:")
print(aggregation)
```

```python
# Filter rows where 'calories' is between 500 and 1000
filtered_df = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]

print("\nFiltered Data:")
print(filtered_df)
```

```python
# Filter rows where 'calories' > 500 and 'pulse' < 100
filtered_df = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]

print("Filtered Data (Calories > 500 and Pulse < 100):")
print(filtered_df)
```

```python
# Create a new DataFrame excluding the 'Maxpulse' column
df_modified = df.drop(columns=['Maxpulse'])

print("\nDataFrame without 'Maxpulse':")
print(df_modified)
```

```python
# Delete the 'Maxpulse' column from the original DataFrame
df.drop(columns=['Maxpulse'], inplace=True)

print("\nOriginal DataFrame after deleting 'Maxpulse':")
print(df)
```

```python
# Convert the 'calories' column to integer datatype
df['Calories'] = df['Calories'].astype(int)

print("\nDataFrame with 'calories' column as integer:")
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration'], df['Calories'], alpha=0.5, edgecolors='w', s=100)

# Add labels and title
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter Plot of Duration vs Calories')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd

# Import the dataset
df = pd.read_csv('Salary_Data.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python

from sklearn.model_selection import train_test_split

# Assuming 'YearsExperience' is the feature and 'Salary' is the target
X = df[['YearsExperience']]
y = df['Salary']

# Split the data into training and testing sets (1/3 test, 2/3 train)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
```

```python
from sklearn.linear_model import LinearRegression

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import matplotlib.pyplot as plt

# Plot training data
plt.figure(figsize=(14, 7))

plt.scatter(X_train, y_train, color='blue', label='Training data')

# Plot test data and predictions
plt.scatter(X_test, y_test, color='red', label='Test data')
plt.scatter(X_test, y_pred, color='green', marker='x', label='Predictions')

# Plotting the regression line
plt.plot(X_train, model.predict(X_train), color='black', linewidth=2, label='Regression line')

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.grid(True)

plt.show()
```
